```python
import os 
import numpy as np
from PIL import Image, ImageTk
import cv2
from tkinter import messagebox
def train_classifier():
    data_dir = ("images")
    path = [os.path.join(data_dir, file) for file in os.listdir(data_dir)]  # list comprehension

    faces = []
    ids = []

    for image in path:
        img = Image.open(image).convert('L')  # Convert into Grayscale image
        imageNp = np.array(img, 'uint8')
        id = int(os.path.split(image)[1].split('.')[1])

        faces.append(imageNp)
        ids.append(id)
        cv2.imshow("Training Photos Sample", imageNp)
        cv2.waitKey(1) == 13
    ids = np.array(ids)  # use for 90% performane increase to covert RMA

    # Train the classifier and save
    face_recognizer = cv2.face.LBPHFaceRecognizer.create()
    face_recognizer.train(faces, ids)
    face_recognizer.write("classifier.xml")
    cv2.destroyAllWindows()
    messagebox.showinfo("Result", "Training Data Sets have been completed!")
train_classifier()
```

```python
import os
import shutil
import cv2
from tkinter import simpledialog, Tk, Frame, Label
from PIL import Image, ImageTk

clf = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
cap = None

def start_capture():
    global lmain, cap, imageFrame

    # Ask user for their name and ID
    root = Tk()
    root.withdraw()
    name = simpledialog.askstring("Input", "Please enter your name:")
    id = simpledialog.askstring("Input", "Please enter your ID:")

    # Set up image directory and file to store names and IDs
    image_dir = os.getcwd() + "/images/"
    if not os.path.exists(image_dir):
        os.mkdir(image_dir)
    names_file = open(os.getcwd() + "/names.txt", "a")

    img_count = 1

    while True:
        img_path = os.path.join(image_dir, f"{name}_{id}_{img_count}.png")
        if not os.path.exists(img_path):
            break
        img_count += 1

    names_file.write(f"{name}_{id}_{img_count}:{name}:{id}\n")
    names_file.close()

    # Set up video capture and display
    imageFrame = Frame(width=1500, height=2000, bd=1, bg='green')
    lmain = Label(imageFrame)
    lmain.grid(row=5, column=10)
    imageFrame.place(relx=.3, rely=.2)
    cap = cv2.VideoCapture(0)

    def temp():
        nonlocal img_count
        _, frame = cap.read()
        cv2image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = clf.detectMultiScale(gray, 1.2, 5)
        for x, y, w, h in faces:
            f = gray[y:y+h, x:x+w]
            cv2.rectangle(cv2image, (x, y), (x+w, y+h), (255, 0, 0), 2)
            cv2.imwrite(img_path, f)
        img_count += 1
        cv2.putText(cv2image, f'Image captured: {img_count}', (10, 22), cv2.FONT_HERSHEY_PLAIN, 2, (255, 255, 0), 2)
        img = Image.fromarray(cv2image)
        imgtk = ImageTk.PhotoImage(image=img)
        lmain.imgtk = imgtk
        lmain.configure(image=imgtk)
        if img_count == 501:
            stop_capture()

        else:
            lmain.after(50, temp) # calls show_frame after 50 milliseconds

    temp()


def stop_capture():
    global cap
    if cap is not None:
        cap.release()
        cap = None


stop_capture()
```
